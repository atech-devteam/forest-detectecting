```python
import os
import sys
import numpy as np
import tensorflow as tf

from pycocotools.coco import COCO
from pycocotools import mask as maskUtils

# Root directory of the project
ROOT_DIR = os.path.abspath("./")
sys.path.append(ROOT_DIR)

from mrcnn import model as modellib, utils
from mrcnn import visualize
from mrcnn.config import Config
from mrcnn.model import log

import forest_detect
```

```python
# Training Configurations
config = forest_detect.DetectConfig()
config.DEFAULT_LOGS_DIR = os.path.join(ROOT_DIR, "logs")
config.GPU_COUNT = 2
config.IMAGES_PER_GPU = 4
config.STEPS_PER_EPOCH = 20
config.LEARNING_RATE = 0.001
config.NUM_CLASSES = 1 + 13 
config.IMAGE_META_SIZE = 1 + 3 + 3 + 4 + 1 + config.NUM_CLASSES 

# datasets, model path
args_model = ROOT_DIR + "/pre_trained_c13.h5"
args_dataset = ROOT_DIR + "/datasets"
args_logs = config.DEFAULT_LOGS_DIR

print("<training configs>")
config.display()
print(" --- path --- ")
print("Model: ", args_model)
print("Dataset: ", args_dataset)
print("Logs: ", args_logs)
```

```python
# Training mode
model = modellib.MaskRCNN(mode="training", config=config, model_dir=args_logs)
model.load_weights(args_model, by_name=True)
```

```python
# Training dataset
dataset_train = forest_detect.ForestDataset()
dataset_train.load_dataset(dataset_dir = args_dataset, subset= "train", class_ids = None)
dataset_train.prepare()  

# Validation dataset
dataset_val = forest_detect.ForestDataset()
dataset_val.load_dataset(dataset_dir = args_dataset, subset= "val", class_ids = None)
dataset_val.prepare()
```

```python
import warnings
warnings.filterwarnings(action='ignore')

print("Training Start!!")
model.train(dataset_train, dataset_val, learning_rate=config.LEARNING_RATE, epochs=2, layers='all', augmentation=False)
print("Training End!!")
```
